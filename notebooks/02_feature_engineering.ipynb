# Notebook 2: Feature Engineering
Build match-level features for IPL prediction


```python
import pandas as pd
import numpy as np
import os
print('Libraries loaded')
```

```python
PROJECT_ROOT = r"D:\IPL 2026 WINNER PREDICTOR"
INTERIM_PATH = os.path.join(PROJECT_ROOT,'data','interim')
PROCESSED_PATH = os.path.join(PROJECT_ROOT,'data','processed')

matches = pd.read_csv(os.path.join(INTERIM_PATH,'cleaned_matches.csv'))
deliveries = pd.read_csv(os.path.join(INTERIM_PATH,'cleaned_deliveries.csv'))

print(matches.shape, deliveries.shape)
```

```python
# Basic team win percentage
team_wins = matches.groupby('winner').size()
total_matches = matches['team1'].value_counts() + matches['team2'].value_counts()
team_win_pct = (team_wins / total_matches).fillna(0).to_dict()

print('Sample team win %:', list(team_win_pct.items())[:5])
```

```python
# Create match-level dataset
rows = []

for _, row in matches.iterrows():
    team1 = row['team1']
    team2 = row['team2']
    winner = row['winner']
    
    rows.append({
        'team1': team1,
        'team2': team2,
        'team1_win_pct': team_win_pct.get(team1,0),
        'team2_win_pct': team_win_pct.get(team2,0),
        'winner': 1 if winner==team1 else 0
    })

model_df = pd.DataFrame(rows)
print(model_df.head())
```

```python
os.makedirs(PROCESSED_PATH, exist_ok=True)
model_df.to_csv(os.path.join(PROCESSED_PATH,'model_ready_data.csv'),index=False)
print('Saved model_ready_data.csv')
```
